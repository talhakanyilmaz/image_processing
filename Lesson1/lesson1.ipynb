```python
m,n,p = im1.shape
m,n,p
```

```python
plt.imshow(im1)
plt.show()
```

```python
newImage = np.zeros((m,n), dtype = int)
newImage
```

```python
for i in range(m):
    for j in range(n):
        s = (im1[i,j,0] + im1[i,j,1] + im1[i,j,2])/3
        newImage[i,j]=s
```

```python
plt.imshow(newImage, cmap = 'gray')
plt.show()
plt.imsave('ITcrowdGray.jpg', newImage, cmap = 'gray')
```

```python
newImage2 = np.zeros((n,m), dtype = int)
newImage2
```

```python
for i in range(m):
    for j in range(n):
        s = (im1[i,j,0] + im1[i,j,1] + im1[i,j,2])/3
        newImage2[j,i]=s
```

```python
plt.imshow(newImage2, cmap = 'gray')
plt.show()
plt.imsave('ITcrowd90.jpg', newImage2, cmap = 'gray')
```
