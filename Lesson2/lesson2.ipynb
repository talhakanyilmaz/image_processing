```python
import matplotlib.pyplot as plt
```

```python
image1 = plt.imread('ITcrowd.jpg')
plt.imshow(image1)
plt.show()
```

```python
image1.ndim, image1.shape
```

```python
m,n,p = image1.shape
image2 = np.zeros((m,n,3), dtype = int)
m,n,image2.shape
for m in range(image1.shape[0]):
    for n in range(image1.shape[1]):
        image2[m,n,0] = image1[m,n,0] - 2
        image2[m,n,1] = image1[m,n,1] + 50
        image2[m,n,2] = image1[m,n,2] - 35
plt.imshow(image2)
plt.show()
```

```python
def renkDegerleriniDegistir(image, s):
    image1 = image
    m,n,p = image1.shape
    image2 = np.zeros((m,n,3), dtype = int)
    for m in range(image1.shape[0]):
        for n in range(image1.shape[1]):
            image2[m,n,0] = image1[m,n,0] - s
            image2[m,n,1] = image1[m,n,1] - s            
            image2[m,n,2] = image1[m,n,2] - s           
    return image
```

```python
def graySatirSutunKes(image):
    m,n,p = image.shape
    new_m = int(m/2)
    new_n = int(n/2)
    image3 = np.zeros((new_m, new_n), dtype = int)
    
    for m in range(new_m):
        for n in range(new_n):
            s = (image[m*2,n*2,0] + image[m*2,n*2,1] + image[m*2,n*2,2])/3
            image3[m,n] = int(s)
    return image3
```

```python
image4 = graySatirSutunKes(image1)
plt.imshow(image4, cmap = 'gray')
plt.show()
plt.imsave('ITcrowdGray.jpg', image4, cmap = 'gray')
```

```python
def renkliSatirSutunKes(image):
    m,n,p = image.shape
    new_m = int(m/2)
    new_n = int(n/2)
    image5 = np.zeros((new_m,new_n,3), dtype = int)
    for m in range(new_m):
        for n in range(new_n):
            image5[m,n,0] = int(image[m*2, n*2, 0])
            image5[m,n,1] = int(image[m*2, n*2, 1])
            image5[m,n,2] = int(image[m*2, n*2, 2])
    return image5
```
