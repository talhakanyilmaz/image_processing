```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = plt.imread('ITcrowd.jpg')
image.shape
```

```python
image2 = np.zeros((1080, 1920), dtype = np.uint8)
image2.shape
image2 = image[:,:,0]
image3 = np.zeros((1080,1920), dtype = np.uint8)
plt.imshow(image2, cmap = 'gray')
plt.show()
```

```python
m,n = image2.shape
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        s = image2[i-1, j-1]/9 + image2[i-1, j]/9 + image2[i-1, j+1]/9 + image2[i, j-1]/9 + image2[i, j]/9 + image2[i, j+1]/9 + image2[i+1, j-1]/9 + image2[i+1, j]/9 + image2[i+1, j+1]/9
        s=int(s)
        image3[i,j] = s               
```

```python
plt.subplot(1,2,2)
plt.imshow(image2, cmap='gray')

plt.subplot(1,2,1)
plt.imshow(image3, cmap='gray')
```

```python
plt.imsave('test1.jpg', image2, cmap='gray')
plt.imsave('test2.jpg', image3, cmap='gray')
```
